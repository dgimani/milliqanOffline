```python
import uproot
```

```python
import time
```

```python
#uproot test
start=time.time()
c=[]
for array in uproot.iterate(
    "/store/user/mcarrigan/trees/v29/MilliQan_Run570.*_v29_firstPedestals.root:t", 
    ["v_groupTDC_g0"]):
    # each of the following have 100 entries
    for a in array['v_groupTDC_g0']:
        print(a)
        c.append(a)

end = time.time()
print("total time:"+str(end-start)) 
#total time 0.9s
```

```python
#check if there are entries missing that leads to less time.
len(c)
```

```python
import ROOT as r
from ROOT import TChain


start=time.time()
treeChain = r.TChain("t")
treeChain.Add('/store/user/mcarrigan/trees/v29/MilliQan_Run570.*_v29_firstPedestals.root')

b=[]
for ievent, event in enumerate(treeChain):
    a = event.v_groupTDC_g0
    print(a)
    b.append(a)

end = time.time()
print("total time:"+str(end-start))
#total time 28s
```
